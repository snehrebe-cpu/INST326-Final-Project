# Frozen Sets

```python
match_lines = [
    "Maryland v Duke",
    "Duke v Maryland",
    "Akron v Penn State",
    "Maryland v Duke",
]

matches = {}
for line in match_lines:
    t1, t2 = line.strip().split(" v ")
    key = frozenset({t1, t2})          
    matches[key] = matches.get(key, 0) + 1

pair = frozenset({"Maryland", "Duke"})
print("Maryland vs Duke games:", matches.get(pair, 0))
```
