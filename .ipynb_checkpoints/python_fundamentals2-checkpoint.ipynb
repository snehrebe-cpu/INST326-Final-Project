## 1) Strings

```
# Basics
s = "University of Maryland"
print(s)                 # whole string
print(len(s))            # length
print(s[0], s[-1])       # first and last char
print(s[0:10])           # slice (0..9)

# Methods
print(s.lower())
print(s.upper())
print(s.replace("Maryland", "Terps"))

# f-strings
name = "Sean"
dept = "Information Science"
msg = f"Hi {name}, welcome to {dept}!"
print(msg)

# Practice: change s and print the last 5 characters.
# TODO: your code here
```

## 2) Lists

```
# Create and access
nums = [10, 20, 30, 40]
animals = ["turtle", "eagle", "terrapin"]
print(nums[0], nums[-1])
print(animals[:2])

# Modify
nums.append(50)
animals[1] = "owl"
print(nums)
print(animals)

# Simple comprehension (build a new list)
doubled = [n * 2 for n in nums]
print(doubled)

# Practice: make a list of the first 5 squares [1, 4, 9, 16, 25].
# TODO: your code here
```

## 3) Loops

```
# for-loop over a list
fruits = ["apple", "banana", "cherry"]
for f in fruits:
    print("fruit:", f)

# for-loop with index
for i, f in enumerate(fruits):
    print(i, f)

# while-loop (count down from 3 to 1)
n = 3
while n > 0:
    print("n =", n)
    n -= 1

# Practice: print only items with length >= 6 from this list.
words = ["campus", "ai", "python", "data", "College Park"]
# TODO: your code here
```

## 4) Files (writing and reading)
**Note:** This writes a small file in the current folder.

```
# Write a file
filename = "sample_notes.txt"
lines = [
    "This is a small text file.\n",
    "Practice makes progress.\n",
    "Go Terps!\n",
]
with open(filename, "w", encoding="utf-8") as f:
    f.writelines(lines)
print("Wrote", filename)

# Read it back
with open(filename, "r", encoding="utf-8") as f:
    text = f.read()
print("--- file contents ---")
print(text)

# Count lines and words (very simple word split)
with open(filename, "r", encoding="utf-8") as f:
    count_lines = 0
    count_words = 0
    for line in f:
        count_lines += 1
        count_words += len(line.split())
print("lines:", count_lines, "words:", count_words)

# Practice: append one more line to the same file, then print the new line count.
# TODO: your code here
```
