```python
import pandas as pd
df06 = pd.read_csv("mens1500_2006.csv")
df10 = pd.read_csv("mens1500_2010.csv")
df14 = pd.read_csv("mens1500_2014.csv")
df18 = pd.read_csv("mens1500_2018.csv")
df22 = pd.read_csv("mens1500_2022.csv")
results = pd.concat([df06, df10, df14, df18, df22], ignore_index=True)
skaters = pd.read_csv("speed_skaters.csv")
df = results.merge(skaters, left_on="Athlete", right_on="Name", how="left")
def time_to_float(t):
    if t != t:
        return t
        m, s = t.split(":")
        return int(m)*60 + float(s)
        df["Seconds"] = df["Result"].apply(time_to_float)
```

```python
I selected only athletes who finished the race, and who skated in 2018 or later. This filters the DataFrame to a focused group. It maintains recent Olympic years and finished performances. Older results and non-finishers are eliminated by the filter. It provides more accurate race time comparisons.
```

```python
I categorized the information by year. Next, I determined each group's average Seconds value. This illustrates how the average time to finish varied throughout the Olympics. The data is divided by year through grouping. Every group is subjected to the same computation by the mean operation. A summary table that displays performance trends over time is created by combining the results.
```
