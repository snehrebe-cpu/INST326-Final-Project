## 1) Strings

```python
# f-strings
name = "Sean"
dept = "Information Science"
msg = f"Hi {name}, welcome to {dept}!"
print(msg)
```

## 2) Lists

```python
# Create and access
nums = [10, 20, 30, 40]
animals = ["turtle", "eagle", "terrapin"]
print(nums[0], nums[-1])
print(animals[:2])

# Modify
nums.append(50)
animals[1] = "owl"
print(nums)
print(animals)

# Simple comprehension (build a new list)
doubled = [n * 2 for n in nums]
print(doubled)

# Practice: make a list of the first 5 squares [1, 4, 9, 16, 25].
# TODO: your code here
```

## 3) Loops

```python
# for-loop over a list
fruits = ["apple", "banana", "cherry"]
for f in fruits:
    print("fruit:", f)

# for-loop with index
for i, f in enumerate(fruits):
    print(i, f)

# while-loop (counts down from 3 to 1)
n = 3
while n > 0:
    print("n =", n)
    n -= 1

```

## 4) Files 

```python
# Write a file
filename = "sample_notes.txt"
lines = [
    "This is a small text file.\n",
    "Practice makes progress.\n",
    "Go Terps!\n",
]
with open(filename, "w", encoding="utf-8") as f:
    f.writelines(lines)
print("Wrote", filename)

# Read it back
with open(filename, "r", encoding="utf-8") as f:
    text = f.read()
print("--- file contents ---")
print(text)

# Counts lines and words
with open(filename, "r", encoding="utf-8") as f:
    count_lines = 0
    count_words = 0
    for line in f:
        count_lines += 1
        count_words += len(line.split())
print("lines:", count_lines, "words:", count_words)
```
